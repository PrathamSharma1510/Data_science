```python
# Q8
import pandas as pd

# Function to calculate the difference between two dates
def calculate_time_difference(date1, date2):
    # Convert strings to pandas Timestamps
    timestamp1 = pd.to_datetime(date1)
    timestamp2 = pd.to_datetime(date2)
    
    # Calculate timedelta
    time_diff = timestamp2 - timestamp1

    # Extracting days, hours and minutes
    days = time_diff.days
    hours, remainder = divmod(time_diff.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    return days, hours, minutes

# Taking user input for dates
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Calculating difference
days, hours, minutes = calculate_time_difference(date1, date2)

print(f"The difference is {days} days, {hours} hours, and {minutes} minutes.")
```
